```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("ted.csv")
df
```

# ratio tedtalks

```python
df['views_like_ratio']=df['views']/df['likes']
sorted=df.sort_values(by='views_like_ratio',ascending=False)
top=sorted.head(2)
top
```

# mostpopularspeaker in term of views

```python
mostpopular=df.groupby('author').views.sum().sort_values(ascending=False).head(1)
print("the most popular ted talks speaker:",mostpopular[0])
```

#  popular tedtalks

```python
ted=df.sort_values(by='views',ascending=False)
ted
```

# mostpopular speaker

```python
speaker=df.author.value_counts()
mostpopular=speaker.idxmax()
mostpopular
```

# month wise tedtalks analysis

```python
df['month']=df['date'].apply(lambda x:x.split('-')[0])
df['year']=df['date'].apply(lambda x:x.split('-')[1])
df
```

```python
import matplotlib.pyplot as plt
month=df.groupby('month').count()
month.plot(kind='bar',y='title',figsize=(5,2))
plt.xlabel('month')
plt.ylabel('frequency')
plt.title('monthly analysis of tedtalks frequency')
plt.show()
```

# year wise ted analysis

```python
import matplotlib.pyplot as plt
year=df.groupby('year').count()
year.plot(kind='bar',y='title',figsize=(5,2))
plt.xlabel('year')
plt.ylabel('frequency')
plt.title('year analysis of tedtalks frequency')
plt.show()
```
