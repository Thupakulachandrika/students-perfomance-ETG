# distribution of a columns

```python
col=list(df.columns)
plt.hist(col)
plt.title('distribution of')
plt.xlabel(col)
plt.ylabel('count')
plt.show()
```

# relationship between colums 

```python
corr=df.corr()
fig=plt.figure()
ax=fig.add_subplot(111)
cax=ax.matshow(corr,cmap='coolwarm',vmin=-1,vmax=1)
fig.colorbar(cax)
ticks=np.arange(0,len(df.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.columns)
plt.show()
```

# rank of all features 

```python
corr=df.corr()
rank=np.argsort(corr,axis=1)
rank
```
